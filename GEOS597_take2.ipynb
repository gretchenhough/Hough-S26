```python
#I need RGB!

def RGB_L89(year, base_path=r"C:\Users\gretchenhough\Documents\ArcGIS\Projects\Bear Lake"):
    folder = os.path.join(base_path, str(year))

    with rasterio.open(os.path.join(folder, f"{year}_B2_scaled_clipped.tif")) as src2:
        b2 = src2.read(1).astype("float32")
        meta = src2.meta.copy()

    with rasterio.open(os.path.join(folder, f"{year}_B3_scaled_clipped.tif")) as src3:
        b3 = src3.read(1).astype("float32")

    with rasterio.open(os.path.join(folder, f"{year}_B4_scaled_clipped.tif")) as src4:
        b4 = src4.read(1).astype("float32")

    # stack as RGB = 3,2,1
    rgb = np.stack([b4, b3, b2])

    # replace nan with 0 before stretch
    rgb = np.nan_to_num(rgb, nan=0.0)

    # simple stretch
    rgb = np.clip(rgb * 3, 0, 1)
    rgb = (rgb * 255).astype(np.uint8)

    # update metadata and REMOVE nodata
    meta.update(
        count=3,
        dtype=rasterio.uint8,
        nodata=None
    )

    out_path = os.path.join(folder, f"{year}_RGB.tif")

    with rasterio.open(out_path, "w", **meta) as dst:
        dst.write(rgb)

    print("Saved:", out_path)
```

```python
import imageio.v2 as imageio
import matplotlib.pyplot as plt

def make_rgb_video(
    years,
    base_path=r"C:\Users\gretchenhough\Documents\ArcGIS\Projects\Bear Lake",
    out_name="BearLake_RGB_timeseries.mp4",
    fps=1.3
):
    frames = []

    for year in years:
        rgb_path = os.path.join(base_path, str(year), f"{year}_RGB.tif")

        with rasterio.open(rgb_path) as src:
            rgb = src.read()  # shape: (3, rows, cols)

        rgb = np.transpose(rgb, (1, 2, 0))  # to (rows, cols, 3)

        fig, ax = plt.subplots(figsize=(8, 8))
        ax.imshow(rgb)
        ax.set_title(str(year), fontsize=18)
        ax.axis("off")

        scalebar = ScaleBar(
        dx=30,              
        units="m",
        location="lower left",
        box_alpha=0.7)
        ax.add_artist(scalebar)

        fig.canvas.draw()
        frame = np.frombuffer(fig.canvas.buffer_rgba(), dtype=np.uint8)
        frame = frame.reshape(fig.canvas.get_width_height()[::-1] + (4,))
        frame = frame[:, :, :3]  # drop alpha

        frames.append(frame)
        plt.close(fig)

        print(f"Added frame: {year}")

    out_path = os.path.join(base_path, out_name)
    imageio.mimsave(out_path, frames, fps=fps)

    print(f"Saved GIF: {out_path}")
```

```python
def indices_L5(year, base_path=r"C:\Users\gretchenhough\Documents\ArcGIS\Projects\Bear Lake"):
    folder = os.path.join(base_path, str(year))
    with rasterio.open(os.path.join(folder, f"{year}_B2_scaled_clipped.tif")) as src2:
        b2 = src2.read(1).astype("float32")
        meta = src2.meta.copy()
    with rasterio.open(os.path.join(folder, f"{year}_B3_scaled_clipped.tif")) as src3:
        b3 = src3.read(1).astype("float32")
    with rasterio.open(os.path.join(folder, f"{year}_B4_scaled_clipped.tif")) as src4:
        b4 = src4.read(1).astype("float32")

    ndwi = (b2 - b4) / (b2 + b4 + 1e-6)
    mask = ndwi > 0
    ndwi_masked = np.where(mask, ndwi, np.nan)
    ndwi_water = mask.astype("uint8")
    ndci = (b4 - b3) / (b4 + b3 + 1e-6)
    ndci_masked = np.where(mask, ndci, np.nan)

    # Save indices as new TIFFs
    meta.update(dtype=rasterio.float32, count=1, nodata=np.nan)
    with rasterio.open(os.path.join(folder, f"{year}_NDWI_raw.tif"), "w", **meta) as dst:
        dst.write(ndwi.astype("float32"), 1)
    with rasterio.open(os.path.join(folder, f"{year}_NDWI.tif"), "w", **meta) as dst:
        dst.write(ndwi_masked.astype("float32"), 1)
    with rasterio.open(os.path.join(folder, f"{year}_NDCI.tif"), "w", **meta) as dst:
        dst.write(ndci_masked.astype("float32"), 1)

    mask_meta = meta.copy()
    mask_meta.update(dtype=rasterio.uint8, count=1, nodata=0)
    with rasterio.open(os.path.join(folder, f"{year}_Water.tif"), "w", **mask_meta) as dst:
        dst.write(ndwi_water, 1)
```

```python
def mask_temp(year, base_path=r"C:\Users\gretchenhough\Documents\ArcGIS\Projects\Bear Lake"):
    folder = os.path.join(base_path, str(year))
    with rasterio.open(os.path.join(folder, f"{year}_T_clipped.tif")) as src:
        temp = src.read(1).astype("float32")
        meta = src.meta.copy()
    with rasterio.open(os.path.join(folder, f"{year}_Water.tif")) as src_water:
        water_mask = src_water.read(1).astype("uint8")

    temp_masked = np.where(water_mask == 1, temp, np.nan)

    meta.update(dtype=rasterio.float32, count=1, nodata=np.nan)
    with rasterio.open(os.path.join(folder, f"{year}_T_water.tif"), "w", **meta) as dst:
        dst.write(temp_masked.astype("float32"), 1)
```

```python
def make_idx_video(
    years,
    base_path=r"C:\Users\gretchenhough\Documents\ArcGIS\Projects\Bear Lake",
    fps=1.3
):
    
    idx=["Water", "NDCI", "T_water"]
    for i in idx:
        frames = []
        if i == "Water":
                cmap = "Blues"
                vmin, vmax = 0, 1
        elif i == "T_water":
                cmap = "inferno"
                vmin, vmax = 12, 25
        elif i == "NDCI":
                cmap = "viridis"
                vmin, vmax = -0.2, 0.5

        for year in years:
            i_path = os.path.join(base_path, str(year), f"{year}_{i}.tif")

            with rasterio.open(i_path) as src:
                index = src.read(1).astype("float32")
            
            


            fig, ax = plt.subplots(figsize=(8, 8))
            im=ax.imshow(index, cmap=cmap, vmin=vmin, vmax=vmax)
            ax.set_title(str(year), fontsize=18)
            ax.axis("off")

            if i == "T_water":
                cbar = fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
                cbar.set_label("Temperature (°C)")
                cbar.ax.tick_params(labelsize=10)
            elif i == "NDCI":
                cbar = fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
                cbar.set_label("NDCI")
                cbar.ax.tick_params(labelsize=10)
                
            scalebar = ScaleBar(
            dx=30,              
            units="m",
            location="lower center",
            box_alpha=0.7)
            ax.add_artist(scalebar)

            fig.canvas.draw()
            frame = np.frombuffer(fig.canvas.buffer_rgba(), dtype=np.uint8)
            frame = frame.reshape(fig.canvas.get_width_height()[::-1] + (4,))
            frame = frame[:, :, :3]  # drop alpha

            frames.append(frame)
            plt.close(fig)

            print(f"Added frame: {year}")
        out_name = f"BearLake_{i}_timeseries.mp4"

        out_path = os.path.join(base_path, out_name)
        imageio.mimsave(out_path, frames, fps=fps)

        print(f"Saved Video: {out_path}")
```

```python
import imageio.v2 as imageio
import matplotlib.pyplot as plt

def make_ndti_video(
    years,
    base_path=r"C:\Users\gretchenhough\Documents\ArcGIS\Projects\Bear Lake",
    out_name="BearLake_NDTI_timeseries.mp4",
    fps=1.3
):
    frames = []
    cmap = "viridis"
    vmin, vmax = -0.2, 0.3

    for year in years:
        ndti_path = os.path.join(base_path, str(year), f"{year}_NDTI.tif")

        with rasterio.open(ndti_path) as src:
            ndti = src.read(1).astype("float32")

        fig, ax = plt.subplots(figsize=(8, 8))
        im=ax.imshow(ndti, cmap=cmap, vmin=vmin, vmax=vmax)
        ax.set_title(str(year), fontsize=18)
        ax.axis("off")

        cbar = fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
        cbar.set_label("NDTI", fontsize=12)
        cbar.ax.tick_params(labelsize=10)

        scalebar = ScaleBar(
        dx=30,              
        units="m",
        location="lower center",
        box_alpha=0.7)
        ax.add_artist(scalebar)

        fig.canvas.draw()
        frame = np.frombuffer(fig.canvas.buffer_rgba(), dtype=np.uint8)
        frame = frame.reshape(fig.canvas.get_width_height()[::-1] + (4,))
        frame = frame[:, :, :3]  # drop alpha

        frames.append(frame)
        plt.close(fig)

        print(f"Added frame: {year}")

    out_path = os.path.join(base_path, out_name)
    imageio.mimsave(out_path, frames, fps=fps)

    print(f"Saved GIF: {out_path}")
```

```python
# Finally ready for some stats! First is surface area.

import pandas as pd


def area(years, base_path=r"C:\Users\gretchenhough\Documents\ArcGIS\Projects\Bear Lake"):
    pixel_area = 30 * 30
    results = []

    for year in years:
        water_path = os.path.join(base_path, str(year), f"{year}_Water.tif")
        with rasterio.open(water_path) as src:
            water=src.read(1)
        water_pixels = np.sum(water == 1)
        area_m2 = water_pixels * pixel_area
        area_km2 = area_m2 / 1e6
        results.append({"Year": year, "Water Pixels": water_pixels, "Area": area_km2})
        print(f"{year}: {area_km2:.2f} km^2")

    df=pd.DataFrame(results)
    return df
        
```

```python
years=[1985, 1986, 1987, 1988, 1989, 1990, 1992, 1993, 1994, 1995, 1996, 1997, 1998, 1999, 2000, 2002, 2003, 2004, 2006, 2007, 2008, 2009, 2010, 2011, 2014, 2015, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025]
df_area=area(years)
```

```python
from scipy.stats import linregress


x=df_area["Year"]
y=df_area["Area"]

result=linregress(x,y)

slope=result.slope
intercept=result.intercept
r2=result.rvalue**2
p=result.pvalue

y_fit=slope*x+intercept

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(df_area["Year"], df_area["Area"], marker="o")
ax.plot(x, y_fit, color="red", label=f"Fit: y={slope:.2f}x+{intercept:.2f}\nR²={r2:.3f}, p={p:.3f}")
ax.set_title("Bear Lake Surface Area Over Time", fontsize=16)
ax.set_xlabel("Year", fontsize=14)
ax.set_ylabel("Area (km^2)", fontsize=14)
plt.text(0.01, 0.6, f"Slope: {slope:.2f}\nR²={r2:.3f}\np={p:.3f}", transform=ax.transAxes, fontsize=14, verticalalignment="top")
plt.show()
```

```python
# Now temp over time:

def temp_stats(years, base_path=r"C:\Users\gretchenhough\Documents\ArcGIS\Projects\Bear Lake"):
    results = []

    for year in years:
        temp_path = os.path.join(base_path, str(year), f"{year}_T_water.tif")
        with rasterio.open(temp_path) as src:
            temp = src.read(1)
        mean_temp = np.nanmean(temp)
        sd_temp = np.nanstd(temp)
        results.append({"Year": year, "Temp": mean_temp, "Temp SD": sd_temp})
        print(f"{year}: {mean_temp:.2f} °C ± {sd_temp:.2f} °C")

    df=pd.DataFrame(results)
    return df
```

```python
years=[1985, 1986, 1987, 1988, 1989, 1990, 1992, 1993, 1994, 1995, 1996, 1997, 1998, 1999, 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011,
       2013, 2014, 2015, 2016, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025]
df_temp=temp_stats(years)
```

```python
import matplotlib.pyplot as plt
from scipy.stats import linregress

x=df_temp["Year"]
y=df_temp["Temp"]

result=linregress(x,y)

slope=result.slope
intercept=result.intercept
r2=result.rvalue**2
p=result.pvalue

y_fit=slope*x+intercept

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(df_temp["Year"], df_temp["Temp"], marker="o")
ax.errorbar(df_temp["Year"], df_temp["Temp"], yerr=df_temp["Temp SD"], fmt="o", color="blue", ecolor="lightblue", capsize=5)
ax.plot(df_temp["Year"], y_fit, color="red", linestyle="--")
ax.set_title("Bear Lake Mean Surface Temperature Over Time", fontsize=16)
ax.set_xlabel("Year", fontsize=12)
ax.set_ylabel("Mean Water Temperature (°C)", fontsize=12)
plt.text(x.min(),y.max(), f"Slope: {slope:.4f} °C/year\nR²: {r2:.4f}\nP-value: {p:.4f}", fontsize=10, verticalalignment="top")
plt.show()
```

```python
def ndti_stats(years, base_path):
    results = []

    for year in years:
        path = os.path.join(base_path, str(year), f"{year}_NDTI.tif")

        with rasterio.open(path) as src:
            ndti = src.read(1).astype("float32")

        valid = ndti[np.isfinite(ndti)]

        results.append({
            "Year": year,
            "MeanNDTI": np.mean(valid),
            "StdNDTI": np.std(valid)
        })

    return pd.DataFrame(results)
```

```python
years=[1985, 1986, 1987, 1988, 1989, 1990, 1992, 1993, 1994, 1995, 1996, 1997, 1998, 1999, 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2018, 2019, 2020, 2021, 2023, 2024, 2025]

df_ndti=ndti_stats(years, r"C:\Users\gretchenhough\Documents\ArcGIS\Projects\Bear Lake")
```

```python
# regression
x = df_ndti["Year"].values
y = df_ndti["MeanNDTI"].values
yerr = df_ndti["StdNDTI"].values

result = linregress(x, y)

slope = result.slope
intercept = result.intercept
r2 = result.rvalue ** 2
p = result.pvalue

y_fit = slope * x + intercept

# plot
plt.figure(figsize=(8, 5))

plt.errorbar(
    x, y,
    yerr=yerr,
    fmt="o-",
    capsize=3,
    label="Mean ± SD"
)

plt.plot(x, y_fit, "--", color="red", label="Best fit")

plt.xlabel("Year")
plt.ylabel("Mean NDTI")
plt.title("Bear Lake Mean NDTI Over Time")

plt.text(
    x.min(),
    y.max()-0.5,
    f"Slope: {slope:.4f} /year\nR²: {r2:.4f}\np-value: {p:.4f}",
    fontsize=11,
    verticalalignment="top"
)

plt.show()

print(f"Slope: {slope:.6f} per year")
print(f"Intercept: {intercept:.4f}")
print(f"R²: {r2:.4f}")
print(f"p-value: {p:.6g}")
print(f"Estimated total change over study period: {slope * (x.max() - x.min()):.4f}")
```
