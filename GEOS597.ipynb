```python
#Let's clean up file names for Landsat 5 (already completed 1985-2000)

def rename_L5(year, folder=r"C:\Users\gretchenhough\Documents\ArcGIS\Projects\Bear Lake"):

    band_map={"_SR_B2": "B2","_SR_B3": "B3","_SR_B4": "B4","_ST_B6": "B6"}

    year_path = os.path.join(folder, str(year))

    for filename in os.listdir(year_path):
        for old_band, new_band in band_map.items():
            if old_band in filename:
                new_filename = f"{year}_{new_band}.tif"
                old_file=os.path.join(year_path,filename)
                new_file=os.path.join(year_path,new_filename)
                os.rename(old_file, new_file)
                print(f"Renamed {filename} to {new_filename}")
                break
```

```python
#Now I need to scale all the files for Landsat 5

def scale_L5(year, base_path=r"C:\Users\gretchenhough\Documents\ArcGIS\Projects\Bear Lake"):
    folder = os.path.join(base_path, str(year))
    bands=[1,2,3,4]
    for b in bands:
        tif_name=f"{year}_B{b}.tif"
        tif_path=os.path.join(folder,tif_name)
        with rasterio.open(tif_path) as src:
            data=src.read(1).astype(np.float32)
            scaled_data=(data*0.0000275)-0.2
            meta=src.meta.copy()
            meta.update(dtype=rasterio.float32)
        out_name=f"{year}_B{b}_scaled.tif"
        out_path=os.path.join(folder,out_name)
        with rasterio.open(out_path, 'w', **meta) as dst:
            dst.write(scaled_data, 1)
    
    with rasterio.open(os.path.join(folder,f"{year}_B2_scaled.tif")) as src:
        data=src.read(1)
        print(data.min(), data.max())
```

```python
# NDWI and NDCI and GNDVI (better for water veg) for Landsat 8/9

def cal_indicesL89(year, base_path=r"C:\Users\gretchenhough\Documents\ArcGIS\Projects\Bear Lake"):
    folder = os.path.join(base_path, str(year))
    with rasterio.open(os.path.join(folder,f"{year}_B3_scaled.tif")) as src:
        b3=src.read(1)
        meta=src.meta.copy()
    with rasterio.open(os.path.join(folder,f"{year}_B4_scaled.tif")) as src:
        b4=src.read(1)
    with rasterio.open(os.path.join(folder,f"{year}_B5_scaled.tif")) as src:
        b5=src.read(1)
        
    ndwi=(b3-b5)/(b3+b5)
    ndci=(b5-b4)/(b4+b5)
    gndvi=(b5-b3)/(b5+b3)
    
    meta.update(dtype=rasterio.float32)
    
    with rasterio.open(os.path.join(folder,f"{year}_NDWI.tif"), 'w', **meta) as dst:
        dst.write(ndwi.astype(np.float32), 1)
    with rasterio.open(os.path.join(folder,f"{year}_NDCI.tif"), 'w', **meta) as dst:
        dst.write(ndci.astype(np.float32), 1)
    with rasterio.open(os.path.join(folder,f"{year}_GNDVI.tif"), 'w', **meta) as dst:
        dst.write(gndvi.astype(np.float32), 1)
```

```python
#I need RGB!

import rasterio
import numpy as np
import os

def RGB_L89(year, base_path=r"C:\Users\gretchenhough\Documents\ArcGIS\Projects\Bear Lake"):
    folder = os.path.join(base_path, str(year))

    with rasterio.open(os.path.join(folder, f"{year}_B4_scaled.tif")) as src4:
        b4 = src4.read(1)
        meta = src4.meta.copy()

    with rasterio.open(os.path.join(folder, f"{year}_B3_scaled.tif")) as src3:
        b3 = src3.read(1)

    with rasterio.open(os.path.join(folder, f"{year}_B2_scaled.tif")) as src2:
        b2 = src2.read(1)

    # Stack RGB
    rgb = np.stack([b4, b3, b2])

    # Stretch to 0–255 (CRITICAL for compatibility)
    rgb = np.clip(rgb * 3, 0, 1)
    rgb = (rgb * 255).astype(np.uint8)

    # FIX metadata properly
    meta.update(
        count=3,
        dtype=rasterio.uint8
    )

    out_path = os.path.join(folder, f"{year}_RGB.tif")

    with rasterio.open(out_path, "w", **meta) as dst:
        dst.write(rgb)

    print("Saved:", out_path)
```

```python
import os
import rasterio
import numpy as np
import imageio.v2 as imageio

def make_rgb_gif(years, base_path=r"C:\Users\gretchenhough\Documents\ArcGIS\Projects\Bear Lake", out_name="bear_lake_rgb.gif", fps=2):

    frames = []

    for y in sorted(years):
        tif_path = os.path.join(base_path, str(y), f"{y}_RGB.tif")

        print("Reading:", tif_path)

        with rasterio.open(tif_path) as src:
            rgb = src.read()  # (bands, rows, cols)

            # Convert to (rows, cols, bands)
            rgb = np.transpose(rgb, (1, 2, 0))

        # If already uint8, this does nothing harmful
        if rgb.dtype != np.uint8:
            rgb = np.clip(rgb * 3, 0, 1)
            rgb = (rgb * 255).astype(np.uint8)

        frames.append(rgb)

    out_path = os.path.join(base_path, out_name)

    imageio.mimsave(out_path, frames, fps=fps)

    print("✅ GIF saved to:", out_path)
```

```python
import rioxarray as rxr
import matplotlib.pyplot as plt
import imageio.v2 as imageio
from pathlib import Path
import numpy as np

# Root folder
root_folder = Path(r"C:\Users\gretchenhough\Documents\ArcGIS\Projects\Bear Lake")
years = [1985, 1986, 1987, 1988, 1989, 1990, 1992, 1993, 1994, 1995, 1996, 1997,
         1998, 1999, 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009,
         2010, 2011, 2013, 2014, 2015, 2016, 2018, 2019, 2020, 2021, 2022, 2023,
         2024, 2025]

indices = ["T", "NDCI", "NDWI", "GNDVI"]

output_img = root_folder / "frames"
output_img.mkdir(exist_ok=True)

FIGSIZE = (8, 6)
DPI = 100


# -----------------------------
# GLOBAL robust scaling (masked pixels only)
# -----------------------------
def compute_global_scale(index_name):
    all_vals = []

    for year in years:
        year_folder = root_folder / str(year)
        tif_files = list(year_folder.glob(f"{year}_{index_name}_lake*.tif"))
        if not tif_files:
            continue

        da = rxr.open_rasterio(tif_files[0], masked=True)
        if da.rio.count > 1:
            da = da[0]

        vals = da.values

        # remove nodata + NaNs
        nodata = da.rio.nodata
        if nodata is not None:
            vals = vals[vals != nodata]

        vals = vals[~np.isnan(vals)]

        if vals.size > 0:
            all_vals.append(vals)

    all_vals = np.concatenate(all_vals)

    # robust range (VERY important for RS data)
    vmin = np.nanpercentile(all_vals, 2)
    vmax = np.nanpercentile(all_vals, 98)

    return float(vmin), float(vmax)


# -----------------------------
# MOVIE MAKER
# -----------------------------
def make_movie(index_name):

    frames = []

    vmin, vmax = compute_global_scale(index_name)
    print(f"{index_name} global scale: vmin={vmin:.3f}, vmax={vmax:.3f}")

    for year in years:
        year_folder = root_folder / str(year)
        tif_files = list(year_folder.glob(f"{year}_{index_name}_lake*.tif"))
        if not tif_files:
            print(f"No file for {year} {index_name}, skipping")
            continue

        da = rxr.open_rasterio(tif_files[0], masked=True)
        if da.rio.count > 1:
            da = da[0]

        fig, ax = plt.subplots(figsize=FIGSIZE, dpi=DPI)

        da.plot(ax=ax, cmap="viridis", vmin=vmin, vmax=vmax)

        ax.set_title(f"{index_name} {year}")
        ax.axis("off")

        img_file = output_img / f"{index_name}_{year}.png"
        fig.savefig(img_file, bbox_inches="tight", pad_inches=0)
        plt.close(fig)

        frames.append(str(img_file))

    if not frames:
        print(f"No frames created for {index_name}")
        return

    gif_path = root_folder / f"{index_name}_movie.gif"
    imageio.mimsave(gif_path, [imageio.imread(f) for f in frames], fps=1)

    print(f"Saved: {gif_path}")


# -----------------------------
# RUN ALL MOVIES
# -----------------------------
for idx in indices:
    make_movie(idx)
```
